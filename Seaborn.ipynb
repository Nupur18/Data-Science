```python
# seaborn is another library for data visualisation and is statistical plotting library
# it is based on top of matplotlib
# it works well with pandas dataframe
import seaborn as sns
import numpy as np
from jupyterthemes import jtplot

jtplot.style(theme='monokai', context='notebook', ticks=True, grid=False)
```

```python
tips = sns.load_dataset('tips')
```

```python
tips
```

```python
tips.head()
```

### Bar Plot

```python
# to create a bar plot
# first it takes 2 things - what should come on x-axis and on y-axis
sns.barplot(x='sex',y='total_bill',data=tips)
# the bars denote the mean bill
```

```python
sns.barplot(x='sex',y='total_bill',data=tips,estimator=np.std)
```

### Count Plot

```python
# aggregation function is always count
sns.countplot(x='sex',data=tips)
```

```python
sns.countplot(x='time',data=tips)
```

### Box Plot

```python
# shows the distribution of categorical data
sns.boxplot(x='day', y='total_bill',data=tips, hue='sex')
# we need to give a categorical value to hue
```

### Violin Plot

```python
sns.violinplot(x='day', y='total_bill',data=tips, hue='sex', split=True)
# split=True combines both distributions together
```

### Distribution Plot

```python
# it allows us to show the distribution of one variable
# we can pick one variable and show its distribution i.e. how spread the variable is
# function only takes one variable
sns.distplot(tips['total_bill'], kde=False, bins=30)
# the line over the graph is called kde (kernel density estimation)
# to make the line invisible, we do kde=False
# we can increase or decrease the number of bins 
```

### KDE Plot

```python
sns.kdeplot(tips['total_bill'])
```

### Joint Plot

```python
# it is like a distribution plot but with two variables
# i.e. we can combine two different distribution plots into one graph
# used to compare two feautures
sns.jointplot(x='total_bill', y='tip', data=tips, kind='scatter')
# kind is used to specify how the data should look like inside the graph
# kind can be scatter,kde,scatter,reg,etc
```

### Pair Plot

```python
# to show a relation between all the features
# it plots pair wise relation between all the feautures of a dataset
sns.pairplot(tips)
```

```python
sns.pairplot(tips, hue='sex')
# create data differently for male and female
```

```python
tips.head()
```

```python
# corr takes non-categorical variables like total_bill, tip and size 
# and calculates correlation between them
tips_corr = tips.corr()
```

```python
tips_corr
```

### Heat Map

```python
# it accepts a panda dataframe that has variables on both rows and columns
sns.heatmap(tips_corr, annot=True)
# to see the numbers we pass annot=True
```

```python
# another technique to create such dataframe that 
# have variables on both rows and columns is pivot table (method in Pandas)
flights = sns.load_dataset('flights')
```

```python
flights
```

```python
flights_pivot = flights.pivot_table(index="month", columns="year", values="passengers")
# index means rows
```

```python
flights_pivot
```

```python
# to visualise flights_pivot as a heatmap
sns.heatmap(flights_pivot)
```

```python
# we can use a different colour map
sns.heatmap(flights_pivot, cmap="coolwarm")
```
