### Line Plot

```python
# to plot a line on a graph
# requirement for plotting these kind of functions is two arrays - x and y (function of x)
x = np.arange(10)
print(x)
y1 = x**2
print(y1)
y2 = 2*x + 3
print(y2)
```

```python
# plotting the two curves on a plot
# first we specify what should go on x-axis, then on y-axis
# marker gives the points through which the line is passing
plt.plot(x,y1,color='red',label='Apple',marker='o')  
# let red line denote price of apple
plt.plot(x,y2,color='green',label='Kiwi',linestyle='dashed',marker='*')   
# let green line denote price of kiwi
plt.xlabel("Time")   # give some title on x-axis
plt.ylabel("Price")   # give some title on y-axis
plt.title("Prices of Fruits over Time")   # give some title to the graph
plt.legend()   # shows the apple and kiwi box
plt.show()
# plt.show() method is used to show the plot
# when we call plot.show(), 
# it means we have finished plotting lines on one graph and are going to next
```

```python
# if we give only 1 array that goes on y-axis and the indices of array go on x-axis
prices = np.array([1,2,3,4])**3
print(prices)
plt.plot(prices)
plt.show()
```

### Scatter Plot

```python
plt.scatter(x,y1)
plt.scatter(x,y2)
plt.show()
# we do not get a line but discrete points
```

```python
# adjust the size of any plot
plt.figure(figsize=(5,5))

# if we do not define a marker in scatter, everything will be circle by default
plt.scatter(x,y1,color='red',label='Apple',marker='^')  
# let red line denote price of apple
plt.scatter(x,y2,color='green',label='Kiwi',linestyle='dashed',marker='*')   
# let green line denote price of kiwi
plt.xlabel("Time")   # give some title on x-axis
plt.ylabel("Price")   # give some title on y-axis
plt.title("Prices of Fruits over Time")   # give some title to the graph
plt.legend()   # shows the apple and kiwi box
plt.show()
```

### Bar Graphs

```python
# tick_label is used to specify what does each item on x-axis denote
x_coordinates = np.array([0,1,2])*2
# we specify x-coordinates, then height of bars, then width
plt.bar(x_coordinates-0.25,[10,20,15],width=0.5,label="Current Year",
       tick_label=["Gold","Silver","Platinum"])   # current year prices
plt.bar(x_coordinates+0.25,[20,10,12],width=0.5,label="Next Year")  # next year prices 
# (0.25 is added to displace 2nd graph from 1st)
plt.title("Metal Prices Comparision")
plt.ylabel("Price")
plt.xlabel("Metal")
plt.legend()
plt.show()
```

```python
# to have a dark background for the graph
plt.style.use("dark_background")
x_coordinates = np.array([0,1,2])*2
plt.bar(x_coordinates-0.25,[10,20,15],width=0.5,label="Current Year",
       tick_label=["Gold","Silver","Platinum"],color="green")   # current year prices
plt.bar(x_coordinates+0.25,[20,10,12],width=0.5,label="Next Year",
        color="orange")   # next year prices 
# (0.5 is added to displace 2nd graph from 1st)
plt.title("Metal Prices Comparision")
plt.ylabel("Price")
plt.xlabel("Metal")
plt.ylim(0,30)    # to put a limit on y, y-axis will have a limit fom 0 to 40
plt.xlim(-2,6)     # to put a limit on x
plt.legend()
plt.show()
```

### Histogram

```python
# to create random data points from standard normal distribution
Xsn = np.random.randn(100)  # we specify how many points we want
print(Xsn)
```

```python
sigma = 8
u = 70
X = Xsn*sigma + u
print(X)
```

```python
# to convert numbers to integers
X1 = np.round(Xsn*sigma + u)
print(X1)
X2 = np.round(Xsn*15 + 40)
print(X2)
```

```python
plt.style.use("seaborn")
plt.hist(X1,alpha=0.8,label="Maths")
plt.hist(X2,alpha=0.9,label="Physics")
# alpha is opacity component
plt.ylabel("Probability/Frequency Count of Students")
plt.xlabel("Marks Range")
plt.title("Histogram")
plt.legend()
plt.show()
```

### Surface Plots
Surface plots are used to :-
- visualise loss functions in Machine Learning and Deep Learning
- visualise state or state value functions in Reinforcement Learning

```python
a = np.array([1,2,3])
b = np.array([4,5,6,7])

# to generate a matrix of points
a,b = np.meshgrid(a,b)
print(a)
print(b)
# vector in a is repeated along the row the number of times as there are elements in b 
# vector in b is repeated along the column the number of times as there are elements in a
# using these two matrices, we can generate a surface on Cartesian plane
```

```python
fig = plt.figure()  # creating a figure
axes = fig.add_subplot(projection="3d")       # axes object comes from matplotlib 3D API
# this gives us a 3d axis
axes.plot_surface(a,b,a+b,cmap="rainbow") # to plot a surface we give what should go on x,y,z axis
# red denotes the region where the value is high
# purple denotes the region where the value is low
plt.show()
# linear surface
```

```python
fig = plt.figure()  # creating a figure
axes = fig.add_subplot(projection="3d")       # axes object comes from matplotlib 3D API
# this gives us a 3d axis
axes.plot_surface(a, b, a**2 + b**2, cmap="rainbow") 
# to plot a surface,we give what should go on x,y,z axis
# red denotes the region where the value is high
# purple denotes the region where the value is low
plt.show()
```

```python
a = np.arange(-1,1,0.02)
b = a

a,b = np.meshgrid(a,b)

fig = plt.figure()  
axes = fig.add_subplot(projection="3d")      
axes.plot_surface(a, b, a**2 + b**2, cmap="rainbow") 
plt.show()
# looks like a bowl shaped surface symmetric about (0,0)
```

### Contour Plots
We cut the surface by parallel planes and the regions where we have the same value of z are denoted by the same colour. 

```python
fig = plt.figure()
axes = fig.add_subplot(projection="3d")      
axes.contour(a, b, a**2 + b**2, cmap="rainbow") 
plt.title("Contour Plot")
plt.show()
```

### Data Visualisation Challenge
Given a movies dataset, the task is to make a visualisation that :-
- plot the lengths of movie title names on x-axis
- frequency counts on y-axis i.e. number of movies having 'x' characters in their title

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
# read a csv file
df = pd.read_csv("movies_metadata.csv", low_memory=False)
```

```python
# to get the first 10 rows
df.head(10)
```

```python
# to view all the columns
df.columns
```

```python
# getting a particular column
# to get all the title in a single list
titles = df.get('title')
```

```python
print(titles)
```

```python
print(type(titles))
```

```python
titles = list(df.get('title'))
print(type(titles))
```

```python
# to view the first 5 titles
print(titles[:5])
```

```python
# we make a dictionary which stores that for a particular length what should be the frequency count
freq_titles = {}

for title in titles:
    length = len(str(title))
    
    if freq_titles.get(length) is None:
        freq_titles[length] = 1
    else:
        freq_titles[length] += 1
```

```python
print(freq_titles)
```

```python
X = np.array(list(freq_titles.keys()))
Y = np.array(list(freq_titles.values()))
```

```python
plt.scatter(X,Y)
plt.xlabel("Length of Movie Title")
plt.ylabel("Frequency Count")
plt.title("Movie Data Visualisation")
plt.show()
```
